```python
## data imgestion
from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt", encoding="utf-8")
text_documents = loader.load()
text_documents
```

```python
#web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(web_path=("https://lilianweng.github.io/posts/2023-06-23-agent/",), bs_kwargs = (dict(parse_only = bs4.SoupStrainer(
    class_ = ("post-title", "post-content","post-header")
))))
documents1 = loader.load()
print(documents1)
print(len(documents1))
```

```python
#pdf loader
from langchain_community.document_loaders import PyPDFLoader
loader1 = PyPDFLoader("apjspeech.pdf")
documents2 = loader1.load()
print(documents2)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from pprint import pprint
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 250)
documents = text_splitter.split_documents(documents2)
pprint(documents[:5])
```

```python
#converting chunks into vector embeddings
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(documents[:20], HuggingFaceBgeEmbeddings(model_name="BAAI/bge-base-en"))
```

```python
#vector database

query = "Whose speech is this?"
result = db.similarity_search(query, k=3)
print(result[0].page_content)
```

```python
#FAISS database
from langchain_community.vectorstores import FAISS
db2 = FAISS.from_documents(documents[:20], HuggingFaceBgeEmbeddings(model_name="BAAI/bge-base-en"))
```

```python
result2 = db2.similarity_search(query, k=3)
print(result2[0].page_content)
```

```python
#vector database

query = "what are the key take aways from this speech?"
retriever = db.as_retriever()
result = retriever.get_relevant_documents(query)
print(result[0].page_content)
```
